##Taking care of missing values

```python
loan.isnull().sum() #finding missing values
```

```python
loan['Gender']=loan['Gender'].fillna(loan['Gender'].mode().iloc[0])
```

```python
loan['Married'] = loan['Married'].fillna(loan['Married'].mode().iloc[0])
```

```python
loan['Dependents'] = loan['Dependents'].fillna(loan['Dependents'].mode().iloc[0])
```

```python
loan['Self_Employed'] = loan['Self_Employed'].fillna(loan['Self_Employed'].mode().iloc[0])
```

```python
loan['LoanAmount'] = loan['LoanAmount'].fillna(loan['LoanAmount'].mean())
```

```python
loan['Loan_Amount_Term'] = loan['Loan_Amount_Term'].fillna(loan['Loan_Amount_Term'].median())
```

```python
loan['Credit_History'] = loan['Credit_History'].fillna(loan['Credit_History'].mean())
```

```python
from sklearn.preprocessing import LabelEncoder
L = LabelEncoder()
```

```python
#label encoding...
loan['Gender']=L.fit_transform(loan['Gender'])
loan['Married']=L.fit_transform(loan['Married'])
loan['Dependents']=L.fit_transform(loan['Dependents'])
loan['Education']=L.fit_transform(loan['Education'])
loan['Self_Employed']=L.fit_transform(loan['Self_Employed'])
loan['Property_Area']=L.fit_transform(loan['Property_Area'])
loan['Loan_Status']=L.fit_transform(loan['Loan_Status'])
```

```python
loan
```

```python
##seperating dependent and independent variables for train_test_split
x = loan.drop(['Loan_Status','Loan_ID'],axis=1)
y = loan['Loan_Status']
```

```python
#importing train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```
