##Splitting The Data

```python
x = data.drop('target',axis=1)
y = data['target']
```

```python
print(x)
```

```python
print(y)
```

##Splitting the data into training data and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=12)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

##Model Training
##Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
model_l = LogisticRegression()
model_l.fit(x_train,y_train)
pred_l = model_l.predict(x_test)
confusion_matrix(y_test,pred_l)
print('Accuracy score on logistic Regression is :',accuracy_score(y_test,pred_l))
```

##Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
model_d = DecisionTreeClassifier()
model_d.fit(x_train,y_train)
pred_d= model_d.predict(x_test)
confusion_matrix(y_test,pred_d)
print('Accuracy score on Decision Tree model is :',accuracy_score(y_test,pred_d))
```

```python
from sklearn import tree
plt.figure(figsize=(25,20))
tree.plot_tree(model_d,filled=True,max_depth=3,fontsize=20)
plt.show()
```

##KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# K = 10
model_k = KNeighborsClassifier(n_neighbors=10)
model_k.fit(x_train,y_train)
pred_k=model_k.predict(x_test)
confusion_matrix(y_test,pred_k)
print('Accuracy score on KNN model when k=10 is :',accuracy_score(y_test,pred_k))
print()
# K = 12
model_k1 = KNeighborsClassifier(n_neighbors=12)
model_k1.fit(x_train,y_train)
pred_k=model_k1.predict(x_test)
confusion_matrix(y_test,pred_k)
print('Accuracy score on KNN model when k=12 is :',accuracy_score(y_test,pred_k))
```

##Support Vector Machine(SVM)

```python
from sklearn.naive_bayes import GaussianNB
model_n = GaussianNB()
model_n.fit(x_train,y_train)
pred_n = model_n.predict(x_test)
confusion_matrix(y_test,pred_n)
print('Accuracy score on Naive Bayes model is :',accuracy_score(y_test,pred_n))
```

##Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
model_n = GaussianNB()
model_n.fit(x_train,y_train)
pred_n = model_n.predict(x_test)
confusion_matrix(y_test,pred_n)
print('Accuracy score on Naive Bayes model is :',accuracy_score(y_test,pred_n))
```

##Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
model_r = RandomForestClassifier()
model_r.fit(x_train,y_train)
pred_r = model_r.predict(x_test)
print('Accuracy score on Random Forest Model is :',accuracy_score(y_test,pred_r))
```
